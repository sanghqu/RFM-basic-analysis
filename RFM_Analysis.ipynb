```python
from datetime import timedelta
date_recency_max=data['InvoiceDate'].max()+ timedelta(days=1)
date_recency_max
```

```python
data['Invoice_Date']=data['InvoiceDate'].dt.date
```

```python
data.head()
```

```python
data_recency= data.groupby(['CustomerID'])['Invoice_Date'].max().reset_index(name='Date_Recency')
data_recency.head()
```

```python
data_recency['Day_recency']=(date_recency_max - pd.to_datetime(data_recency['Date_Recency'])).dt.days
data_recency
```

```python
frequency_buy = data.groupby(['CustomerID'])['InvoiceNo'].count().reset_index(name='frequency_buy')
```

```python
frequency_buy.head()
```

```python
data[data['CustomerID']=='536365']
```

```python
data['Total_unitprice']=data['Quantity']*data['UnitPrice']
```

```python
data.head()
```

```python
monetary_buy= data.groupby(['CustomerID'])['Total_unitprice'].sum().reset_index(name='Monetary_buy')
monetary_buy.head()
```

```python
df1= data_recency.merge(frequency_buy, how='inner', on='CustomerID')
```

```python
final_data=df1.merge(monetary_buy, how='inner', on='CustomerID')
final_data.head()
```

```python
final_data['Recency']=pd.qcut(final_data['Day_recency'],5,labels=[5,4,3,2,1])
```

```python
final_data['frequency_buy'].describe([0.01,0.05,0.1,0.3,0.5,0.6,0.7,0.8,0.9])
```

```python
final_data['Frequency']=pd.qcut(final_data['frequency_buy'],5, labels=[1,2,3,4,5])
```

```python
final_data['Frequency'].value_counts()
```

```python
final_data[final_data['Frequency']==1].head(100)
```

```python
final_data['Monetary']=pd.qcut(final_data['Monetary_buy'],5, labels=[1,2,3,4,5])
```

```python
final_data.head()
```

```python
def Score(x):
    return str(x['Recency'])+ str(x['Frequency'])+str(x['Monetary'])

final_data['RFM']=final_data.apply(Score, axis=1)
final_data
```

```python
final_data['Score']=final_data[['Recency','Frequency','Monetary']].sum(axis=1)
```

```python
final_data.head(100)
```

```python
final_data.head()
```

```python
final_data['Score'].value_counts()
```

```python
def segment_levels(data):
    if data['Score']>=12:
        return 'Cant Lose'
    elif ((data['Score']>=10)&(data['Score']<12)):
        return 'Champition'
    elif ((data['Score']>=9) & (data['Score']<10)):
        return 'Loyal'
    elif ((data['Score'] >= 8) and (data['Score'] < 9)):
        return 'Potential'
    elif ((data['Score'] >= 7) and (data['Score'] < 8)):
        return 'Promising'
    elif ((data['Score'] >= 6) and (data['Score'] < 7)):
        return 'Needs Attention'
    else:
        return 'Require Activation'
    
    
```

```python
final_data['segement_levels']=final_data.apply(segement_levels, axis=1)
final_data.head()
```

```python
final_data['Monetary']=final_data['Monetary']
```

```python
final_data['segment_levels'].nunique()
```

```python
RFM=final_data.groupby('segment_levels').agg({
    'Day_recency':'mean',
    'frequency_buy':'mean',
    'Monetary_buy':['mean','count']
}).round(1)
RFM
```

```python
import matplotlib.pyplot as plt
import squarify
RFM.columns = RFM.columns.droplevel()
RFM.columns = ['RecencyMean','FrequencyMean','MonetaryMean', 'Count']
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(10, 8)
squarify.plot(sizes=RFM['Count'], 
              label=['Can\'t Loose Them',
                     'Champions',
                     'Loyal',
                     'Needs Attention',
                     'Potential', 
                     'Promising', 
                     'Require Activation'], alpha=0.8 )
plt.title("RFM Segments",fontsize=15,fontweight="bold")
plt.axis('off')
plt.show()
```
